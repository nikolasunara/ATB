### Zadatak 1

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
i = 3

while 1:
    i += 1
    if np.log( i ) ** 13 < i ** 3:
        print( i - 1 )
        break
```

```python
fig, ax = plt.subplots(figsize=(10,5))

t1 = np.arange(1, 30000, 1.0)
t2 = np.log( t1 ) ** 13
t3 = np.arange(1, 30000, 1.0)
t4 = pow(t3, 3)

ax.plot(t1, t2, label="ln")
ax.plot(t3, t4, label= "3. potencija")


plt.legend(loc="upper left")
plt.show()
```

Rješenje je **21181**.

### Zadatak 2

```python
import math
```

```python
n = 1.0
i = 0

while 1:
    i += 1
    if math.exp( pow( np.log(n), 1/3) ) < pow( n, 1/16):
        print( i )
        break
    n *= 10
```

Najveći broj za koji vrijedi uvjet zadatka ima **28 znamenaka**.
